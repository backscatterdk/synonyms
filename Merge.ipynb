```python
def merge(lst):
    out = []
    for w in lst:
        try:
            out.append(merge_dic.loc[w]['terms'])
        except:
            out.append(w)
    return out


data = pd.read_csv('nouns_output.csv')

data['merged'] = data['filtered'].apply(
    lambda x: x.split(', ') if not pd.isnull(x) else [])

data['merged'] = data['merged'].apply(merge)

print (d)

data.head()
```
